# Laberindos

```python
from ipythonblocks import BlockGrid
```

```python
def obtener_ancho_largo(filepath):
    width = 0
    with open(filepath, 'r') as file:
        width = len(file.readline())
    height = 0
    with open(filepath, 'r') as file:
        for line in file:
            print(line, end="")
            height += 1
    return (width, height)
```

```python
def generar_laberindo(filepath):
    """Mapea el labarindo ascii en un objeto BlockGrid"""
    with open(filepath, 'r') as file:
        width, height = obtener_ancho_largo(filepath)
        grid = BlockGrid(width-1, height, fill=(225, 225, 225), block_size=20)
        for i, line in enumerate(file):
            for character, block in zip(line, grid[i]):
                if character == '#':
                    block.rgb = (0, 0, 0)
    return grid
```

```python
laberindo = generar_laberindo('laberindo1.txt')
laberindo
```
