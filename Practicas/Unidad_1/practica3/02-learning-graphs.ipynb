```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

## Crear una gr√°fica que forme un circulo

```python
# 9->0->1->2->3->4->5->6
n = 10
circulo = nx.Graph()
nodes = range(n)

for i in range(n):
    circulo.add_edge(i, (i+1) % n)

nx.draw(circulo, with_labels=True)
```

## Construccion de una graficas random
Sirve para probar nuestro algoritmos.

```python
n = 10
p = 0.5  # probabilida de que este conectado
g = nx.gnp_random_graph(n, p)
nx.draw(g)
```

## Construccion de un arbol aleatorio
Cada par de vertice esta conectado a lo mucho por una arista

```python
n = 15
g = nx.random_tree(n)
nx.draw(g)
```

## Construccino de una grafica completa
Cada nodo esta conectada con las demas

```python
n = 10 
g = nx.complete_graph(10)
nx.draw(g)
```
