```python
# Importamos las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt

import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_node('A')
G.add_node((2, 3))
G.add_nodes_from([5, 3.2, 'B', 'C'])
```

```python
G.add_node('P', ancho=3, largo=5)
```

```python
G.nodes
```

```python
# Acceso a un nodo especifico (like dictionaries)
# Por esa misma razon los vertices tiene ser hash object
G.nodes[1]
```

```python
G.nodes['P']
```

```python
# help(nx.draw)
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.nodes
```

```python
# Aristas
G.add_edge(1, 'B')
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.add_edge(88, 'B')
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.add_edge(1, 5, weight=5)
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
G.add_weighted_edges_from([('A', 'P', 55), ('C', 'P', 32)])
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```
