```python
# Importamos nuestra libreria para networks
import networkx as nx

# Necesitaresmo de librerias auxialares para graficar el grafo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Empty graph with no nodes and no edges
G = nx.Graph()
```

Nodes can be any hashable object e.g., a text string, an image, an XML object, another Graph, a customized node object, etc.

```python
# One node at a time
G.add_node(1)
# Any iterable container
# list
# dicts
# tuples
G.add_nodes_from([2,3])
```

You can also add nodes along with node attributes if your container yields 2-tuples 

```python
G.add_nodes_from([
    (4, {'color': 'red'}),
    (5, {'color': 'green'}),
])
```

```python
G.nodes
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

## Creando una grafica no dirigida

```python
nodes = range(10)
graph = nx.Graph()
graph.add_nodes_from(nodes)
graph
```

```python
# graph.nodes[2]  # Accder a un elemento de la grafica
graph.nodes
```

```python
nx.draw(graph, with_labels=True, font_weight='bold')
```

## Agregando aristas a la grafica

```python
edges = [(3, 7), (9, 8), (6, 5), (0, 8), (6, 1)]
graph.add_edges_from(edges)
```

```python
# graph.edges[(3,7)] Accediento a una arista
graph.edges
```

```python
nx.draw(graph, with_labels=True)
```

## Creando una grafica atributos en los nodos (color)

```python
data = range(10)
nodes = zip(
    data,
    [dict(color='green') for _ in range(10)]
)
```

```python
G = nx.Graph()
G.add_nodes_from(list(nodes))
```

```python
G.nodes
```

```python
nx.draw(G, with_labels=True)
```

## Aristas con peso

```python
edges_weights = [(8, 7, 10), (0, 5, 43), (4, 2, 23),
                 (9, 8, 12), (2, 6, 99)]
```

```python
G.add_weighted_edges_from(edges_weights)
```

```python
G.edges[0, 5]
```

```python
nx.draw(G, with_labels=True)
```

```python
# Layout de nuestra grafica
pos = nx.spiral_layout(G)
```

```python
nx.draw(G, pos=pos)
```

## Layouts

```python
G = nx.path_graph(15)
nx.draw(G, with_labels=True)
```

```python
G = nx.gnp_random_graph(15, 0.12)
pos = nx.spring_layout(G)
nx.draw(G, pos=pos, with_labels=True)
```

```python
G = nx.gnp_random_graph(30, 0.12)
pos = nx.spiral_layout(G)
nx.draw(G, pos=pos, with_labels=True)
```

```python
G = nx.gnp_random_graph(30, 0.12)
pos = nx.circular_layout(G)
nx.draw(G, pos=pos, with_labels=True)
```

```python
G = nx.gnp_random_graph(30, 0.12)
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos=pos, with_labels=True)
```
